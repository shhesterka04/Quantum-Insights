# **the Graph Isomorphism problem**

# Использование QPE

Один из возможных подходов с использованием QPE состоит в том, чтобы сначала преобразовать матрицы смежности двух графов в диагональную форму с помощью квантового преобразования Фурье (QFT), а затем применить QPE для извлечения собственных значений диагональных матриц. Сравнивая полученные собственные значения, можно определить, изоморфны графы или нет.



# Использование concept of "quantum similarity".

**Идея:** определить меру квантового сходства между двумя графами, которая фиксирует степень структурного сходства между ними. Эта мера может быть основана на спектральных свойствах матриц смежности графов или на других инвариантах теории графов.

После можно использовать методы квантового машинного обучения и квантовой оптимизации для поиска изоморфизма графов между двумя графами:

* квантовые алгоритмы для нахождения максимального разреза графа, чтобы найти 
биекцию между вершинами двух графов, которая максимизирует их сходство

* квантовые алгоритмы для вариационной оптимизации, чтобы найти оптимальную биекцию, которая максимизирует меру сходства.

**Плюсы данного подхода:** может быть более эффективным и надежным, чем предыдущие квантовые подходы, основанные на QPE или квантовых блужданиях, поскольку опирается на более гибкую и общую меру сходства графов, которая может охватывать широкий спектр структурных сходств между двумя графами. 

**применение раньше:**
Концепция «квантового подобия» использовалась в некоторых предыдущих работах для изучения различных проблем с графами, однако, насколько гпт известно, ранее он не был широко изучен как метод решения проблемы изоморфизма графов.

Одной из актуальных работ в этом направлении является недавняя статья Дуана, Цзяна и Лу (2021), в которой они предлагают квантовый алгоритм изоморфизма графов, основанный на мере квантового подобия, определенной с использованием квантовой информации Фишера. Однако их алгоритм по-прежнему ограничен конкретным классом графов с определенными симметриями.

2. Разработать более эффективный оракул. Основным узким местом в алгоритме, предложенном Дуаном, Цзяном и Лу, является оракул, вычисляющий квантовую информационную матрицу Фишера.

3. Использовать классическую предварительную обработку (можно использовать классические алгоритмы для поиска определенных инвариантов или подграфов графа, которые могут помочь ограничить пространство поиска для квантового алгоритма).

4. Изучить альтернативные квантовые методы (квантовые блуждания, усиление амплитуды или вариационные квантовые алгоритмы).
