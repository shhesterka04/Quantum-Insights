# **Scalable Quantum Algorithm for Solving Linear Systems with Sparse Matrices**

Одним из таких квантовых алгоритмов для решения линейных систем с разреженными матрицами является алгоритм HHL (названный в честь его изобретателей Харроу, Хассидима и Ллойда). 

Основная идея алгоритма HHL состоит в том, чтобы использовать квантовое преобразование Фурье для преобразования входного вектора в квантовое состояние, применить унитарную матрицу, которая кодирует матрицу A, а затем использовать другое квантовое преобразование Фурье для преобразования результирующего состояния обратно в классическое домен. Затем алгоритм измеряет результирующее состояние, чтобы получить вектор решения.
