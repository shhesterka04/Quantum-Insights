# *Пункт 1:*

***Смысл задачи:***

Задача состоит в том, чтобы найти непрерывный подмассив с наибольшей суммой по заданному массиву чисел. Более формально, если у нас есть массив из n чисел a[1], a[2], ..., a[n], максимальный подмассив — это непрерывная подпоследовательность a[i], a[i+1], . .., a[j] такой, что сумма его элементов максимальна среди всех возможных смежных подпоследовательностей.

***Практическое применение:***

Проблема максимального подмассива имеет важные практические приложения, включая обработку сигналов, компьютерное зрение и финансы. Одно известное приложение находится в области анализа фондового рынка, где можно использовать задачу о максимальном подмассиве, чтобы найти наиболее выгодное время для покупки и продажи акций.

***Сложность задачи:***

Сложность задачи о максимальном подмассиве зависит от алгоритма, используемого для ее решения. Подход грубой силы к этой проблеме включает в себя проверку всех возможных непрерывных подпоследовательностей массива, что занимает O (n ^ 2) времени. Однако для решения этой проблемы существует хорошо известный алгоритм «разделяй и властвуй», время выполнения которого лучше O(n*log n).

***NP сложная?***

Определение того, является ли задача максимального подмассива NP-сложной, является открытым исследовательским вопросом. Однако существует родственная проблема, называемая проблемой максимального параллельного подмассива, которая, как известно, является NP-трудной. Эта задача заключается в нахождении двух непересекающихся подмассивов массива, которые имеют наибольшую сумму одновременно. Этот результат предполагает, что задача максимального подмассива также, вероятно, будет NP-сложной. Тем не менее, для задачи о максимальном подмассиве существуют полиномиальные алгоритмы, и многие исследователи считают, что эта задача не является NP-трудной.

# *Пункт 2:*

***наиболее часто используемые классические алгоритмы:***


***Brute Force Algorithm:*** этот алгоритм проверяет все возможные непрерывные подпоследовательности массива и находит ту, у которой максимальная сумма. Как упоминалось ранее, его временная сложность составляет **O(n^2)**.

***Divide and Conquer Algorithm:*** этот алгоритм делит массив на две половины, рекурсивно находит максимальный подмассив в каждой половине, а затем объединяет два решения, чтобы найти максимальный подмассив всего массива. Он имеет временную сложность O(n * log n). 

***Kadane's Algorithm:*** этот алгоритм представляет собой эффективное решение для динамического программирования, которое сканирует массив слева направо, отслеживая максимальную сумму, наблюдаемую до сих пор, и максимальную сумму, заканчивающуюся в каждой позиции в массиве. Он имеет временную сложность **O(n)**.

***Enumeration algorithm:*** Этот алгоритм генерирует все возможные подмассивы данного массива и находит тот, у которого максимальная сумма. Он имеет временную сложность **O (2 ^ n)**.

***Linear Time Algorithm:*** этот алгоритм сканирует массив слева направо, отслеживая максимальную сумму, просмотренную до сих пор, а также начальный и конечный индексы максимального подмассива, просмотренного до сих пор. Он имеет временную сложность **O(n)**.


***Сильные и слабые стороны:***

1. ***Brute force algorithm:***
     - Сильные стороны: простой и понятный. Гарантия нахождения правильного решения.
     - Недостатки: очень медленный для больших массивов.

2. ***Algorithm "divide and conquer":***
     - Сильные стороны: быстрее, чем алгоритм грубой силы, на практике более эффективен для больших массивов.
     - Слабые стороны: требуется более сложная реализация и дополнительное место в памяти для рекурсии. Не всегда может дать оптимальное решение для некоторых крайних случаев.

3. ***Kadane algorithm:***
     - Сильные стороны: быстрый и эффективный, требуется только дополнительное постоянное пространство памяти.
     - Слабые стороны: возвращает только максимальную сумму подмассива, а не сам подмассив. Не подходит для случаев, когда максимальный подмассив может быть пустым.

4. ***Transfer algorithm:***
     - Сильные стороны: простой и понятный. Гарантия нахождения правильного решения.
     - Слабые стороны: Очень медленный для больших массивов, не подходит для практического применения.

5. ***Linear time algorithm:***     
     - Сильные стороны: быстрый и эффективный, возвращает как максимальную сумму подмассива, так и сам подмассив.
     - Слабые стороны: требуется более сложная реализация и дополнительное пространство памяти для отслеживания индексов подмассивов.

Важно отметить, что сильные и слабые стороны каждого алгоритма зависят от конкретного экземпляра задачи и размера массива.

$$x_i + x_{i+1} + ... + x_j = j - i + 1$$

где $i$ и $j$ — индексы первого и последнего элементов подмассива соответственно. Это ограничение гарантирует, что все элементы в подмассиве являются смежными.

```python
!pip install qiskit 
!pip install qiskit.aqua.operators
import numpy as np
from qiskit import Aer, QuantumCircuit
from qiskit.aqua.operators import PauliSumOp

def create_max_subarray_hamiltonian(w):
    n = len(w)
    pauli_list = []
    
    for i in range(n):
        for j in range(i, n):
            if i == j:
                pauli_list.append([1 - w[i][i], 'I'*i + 'Z' + 'I'*(n-i-1)])
            else:
                pauli_list.append([-w[i][j], 'I'*i + 'Z' + 'I'*(j-i-1) + 'Z' + 'I'*(n-j-1)])
                pauli_list.append([-w[i][j], 'I'*i + 'Z' + 'I'*(j-i-1) + 'Z' + 'I'*(n-j-1)])
    
    return PauliSumOp.from_list(pauli_list)
```

***Описание к коду***

Функция create_max_subarray_hamiltonian принимает входную матрицу весов w и генерирует соответствующий гамильтониан как объект PauliSumOp. Гамильтониан строится с использованием операторов Паули I, X, Y и Z, где X и Y не нужны для этой конкретной задачи. Полученный гамильтониан представляет собой отрицательную сумму всех членов внутреннего цикла, где каждый член является произведением Z-операторов на соответствующих кубитах.

Обратите внимание, что этот код предполагает, что веса во входной матрице w неотрицательны. Если есть отрицательные веса, их можно сдвинуть, добавив константу ко всем весам.

Вот код для определения оператора QAOA для задачи максимального подмассива в Qiskit:

```python
from qiskit import QuantumCircuit, QuantumRegister
from qiskit.opflow import PauliSumOp
from qiskit.circuit.library import XGate

def qaoa_operator(gammas, betas, hamiltonian, mixer=XGate()):
    """Defines the QAOA operator for the given Hamiltonian and mixer"""
    
    # Create a quantum register for the qubits
    q = QuantumRegister(hamiltonian.num_qubits)
    qc = QuantumCircuit(q)
    
    # Apply the mixing Hamiltonian to each qubit
    qc.append(mixer, q)
    
    # Apply the cost Hamiltonian to the circuit
    cost_op = hamiltonian.to_opflow().to_paulisum_op()
    qc += cost_op.evolve(evo_time=gammas, num_time_slices=len(gammas)).to_circuit()
    
    # Repeat the mixing Hamiltonian
    qc.append(mixer, q)
    
    # Rotate the qubits to the X basis
    qc.h(q)
    
    # Apply the final rotation around the X-axis
    qc.rx(betas[-1], q[-1])
    
    # Return the quantum circuit
    return qc
```

Здесь гамма и бета — это параметры, которые должны быть оптимизированы классическим оптимизатором. hamiltonian — это гамильтониан для задачи максимального подмассива, выраженного в виде объекта PauliSumOp, а mix — это гамильтониан смешивания, который по умолчанию использует оператор XGate(). Функция qaoa_operator возвращает объект QuantumCircuit, представляющий оператор QAOA для заданных параметров.

```python
import numpy as np
from qiskit import Aer, execute
from qiskit.circuit.library import RealAmplitudes
from qiskit.aqua.operators import WeightedPauliOperator
from scipy.optimize import minimize


def max_subarray_qaoa(weights):
    n = len(weights)
    # Determine the number of qubits needed to represent the problem
    num_qubits = n
    
    # Code the binary optimization problem as a Hamiltonian
    pauli_list = []
    for i in range(n):
        for j in range(i+1, n):
            wp = -weights[i][j]*np.kron(np.eye(2**(n-j-1)), np.kron(np.kron(np.array([[0,1],[0,0]]), np.eye(2**(j-i-1))), np.kron(np.kron(np.array([[0,1],[0,0]]), np.eye(2**(i))), np.eye(2**(n-i-1))))
            pauli_list.append([wp, 1.0])
    operator = WeightedPauliOperator(paulis=pauli_list)
    
    # Create a quantum circuit that prepares the initial state for the QAOA algorithm
    circuit = RealAmplitudes(num_qubits)
    
    # Define the QAOA operator as the sum of the cost and mixer Hamiltonians
    def qaoa_operator(gamma, beta):
        return operator.evolve(
            evo_time=gamma,
            num_time_slices=1,
            quantum_registers=circuit.qregs[0]
        ) @ circuit.decompose() @ operator.evolve(
            evo_time=beta,
            num_time_slices=1,
            quantum_registers=circuit.qregs[0]
        )
    
    # Set the depth of the QAOA algorithm
    depth = 2
    
    # Define the objective function to be minimized
    def objective_function(params):
        gamma = params[:depth]
        beta = params[depth:]
        qc = qaoa_operator(gamma, beta)
        result = execute(qc, Aer.get_backend('qasm_simulator')).result()
        counts = result.get_counts(qc)
        expectation = 0.0
        for bit_string in counts:
            probability = counts[bit_string]/1024
            state = np.array([int(bit) for bit in reversed(bit_string)])
            expectation += probability * operator.expectation_value(state)
        return expectation

    # Minimize the objective function using classical optimization
    initial_params = np.random.uniform(0, np.pi, 2*depth)
    result = minimize(objective_function, initial_params, method='COBYLA')
    optimal_params = result.x
    
    # Calculate the maximum subarray using the optimal parameters
    qc = qaoa_operator(optimal_params[:depth], optimal_params[depth:])
    result = execute(qc, Aer.get_backend('qasm_simulator')).result()
    counts = result.get_counts(qc)
    max_count = 0
    max_subarray = None
    for bit_string in counts:
        state = np.array([int(bit) for bit in reversed(bit_string)])
        count = counts[bit_string]
        subarray = np.where(state == 1)[0]
        subarray_sum = np.sum([weights[i][j] for i in subarray for j in subarray if j > i])
        if subarray_sum > max_count:
            max_count = subarray_sum
            max_subarray = subarray
    
    return max_subarray
```
